```python
import pandas_datareader.data as web
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Data_loader function

```python
def yahoo_data_loader(ticker):
    
    data = web.get_data_yahoo(ticker, start='2004.01.01', end='2020.12.31')
    data.rename(columns = {'Adj Close':ticker}, inplace=True)
    data = data[str(ticker)]
    
    plt.figure(figsize=(14,6))
    plt.title(ticker+ ' History', fontsize=18)
    plt.plot(data)
    plt.xlabel('Date', fontsize=18)
    plt.ylabel(ticker, fontsize=18)
    plt.show()
    
    
    return data
```

```python
def fred_data_loader(ticker):
    data = web.DataReader([ticker],'fred',start='2004.01.01', end='2020.12.31')
    data.rename(columns = {'Adj Close':ticker}, inplace=True)
#   data.rename(columns = {'DATE':'Date'}, inplace=True)
    data = data[str(ticker)]
    
    plt.figure(figsize=(14,6))
    plt.title(ticker+ ' History', fontsize=18)
    plt.plot(data)
    plt.xlabel('Date', fontsize=18)
    plt.ylabel(ticker, fontsize=18)
    plt.show()
    
    
    return data
```

# Merge Datasets

### Join data. Retain all values and all rows, including Null.

```python
df_x_list= [vix, cpi, M2, FEDFUNDS, spread_10to2, Aaa, Unemployment, PPI, USPHCI, IQ]

final_df = sp500
for i in df_x_list:
    final_df = pd.concat([final_df, i], axis=1)
    
final_df
```
